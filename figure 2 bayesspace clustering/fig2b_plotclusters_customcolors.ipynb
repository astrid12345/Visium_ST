```r
rm(list = ls())

# load libraries
library(dplyr)
library(BayesSpace)
library(patchwork)
library(ggplot2)
library(RColorBrewer)

indir = "<yourindir>"
outdir = "<youroutdir>"


# Function to plot clusters with customized lay-out
# 
# Args:
#   sce: SingleCellExperiment object containing clustering results.
#   lesiontype: A string indicating the type of lesion for the title.
#   sample_ID: A string indicating the sample ID for the title.
#   cluster_cols: A vector of colors for the clusters.
#
# Returns:
#   A ggplot object representing the cluster plot.

plot_clusters <- function(sce, lesiontype, sample_ID, cluster_cols){
  plot <- BayesSpace::clusterPlot(sce, palette = cluster_cols, color = NA) # original clusters
  plot <- plot  +  labs(title=paste(sampleID, lesiontype)) 
  return(plot)
} 
 
# example code runs for two control samples.
sampleIDs =  c("ST31", "ST32")
# corresponding example q-values for two control samples.
sampleQs <- c(2, 3) 

# define your custom colors.
cluster_colors = list("ST31" = c("#004d00",
                                 "#669466"),
                      "ST32" = c("#004d00"))

# read in data sample by sample, plot clusters with custom colors, store result in outdir
for (i in 1:length(sampleIDs)){
    sampleID <- sampleIDs[i]
    q <- sampleQs[i]
    clustercol <- unlist(cluster_colors[i], use.names = FALSE)
    print(clustercol)
    print(sampleID)
    
    sce <- readRDS( file=paste0(indir, sampleID, "_sce_q", q,".rds"))
    
    p1 <- plot_clusters(sce, lesiontype = "control", sample_ID = sampleID, cluster_cols = clustercol)
    
    pdf(file = paste0(outdir, sampleID, "_clusterplot_fancycolors.pdf"), width = 5, height = 5)
    print(p1)
    dev.off()
    print(p1)
    }
```
