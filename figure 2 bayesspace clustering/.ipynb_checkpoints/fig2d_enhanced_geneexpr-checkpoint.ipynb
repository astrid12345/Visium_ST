```r
rm(list = ls())
library(BayesSpace)
library(RColorBrewer)
library(patchwork)
library(future)
library(assertthat)
library(xgboost)
library(ggplot2)
indir = "/<yourindir>/"
```

###  optional formatting function ###

```r
#' Plot Expression Data with Custom Titles and Colors
#'
#' This function formats a feature plot of single-cell expression data with custom titles and colors.
#'
#' @param sce A SingleCellExperiment object. 
#' @param gene Character. The gene to be plotted.
#' @param title Character. The title of the enhanced gene expression plot.
#' @param expr_limits Numeric vector of length 2. The limits for the expression scale. Optional.
#'
#' @return A ggplot object representing the feature plot.

#' @example plot_expression(sce.enhanced, gene = "APOC1", title = "ST33_activelesion", expr_limits = c(0,5))

plot_expression <- function(sce, gene, title, expr_limits = NULL) {
    featurePlot(sce, gene, color=NA) +
        scale_fill_gradientn(colors = my_cols, limits = expr_limits, na.value = "#000000") + 
        labs(title=title, fill="Log-norm\nexpression")
}
```

### User inputs

```r
# marker gene to enhance
gene_of_interest <- c("APOC1") 

# optional,the limits for the color scale indicating expression.
gene_limits <- c(0, 5)

# samples of interest to plot
sample <- c("ST33")

# color scale for enhanced gene expression
my_cols <- brewer.pal(n = 9, name = "Greys")  
```

```r
# Load the enhanced single-cell experiment data and plot the expression of a gene of interest

# Load the enhanced dataset for the sample
print(paste("loading", sample))
sce.enhanced <- readRDS(paste0(indir, sample, "_enhanced.rds"))

# Try to get logcounts for the gene of interest
count_for_gene <- try(logcounts(sce.enhanced[(rownames(sce.enhanced) %in% gene_of_interest), ]), silent = TRUE)

if (class(count_for_gene) == "try-error") {
    # If gene data is not available, notify and enhance the features
    print(paste(gene_of_interest, "not yet imputed for enhanced resolution"))
    sce <- readRDS(file = Sys.glob(file.path(paste0(indir, sample, '_sce*.rds')))) # Load original single-cell experiment data

    print('calculating enhanced marker gene expression')
    # Enhance features using xgboost model
    sce.enhanced <- BayesSpace::enhanceFeatures(sce.enhanced, sce,
                                                model = "xgboost",
                                                feature_names = gene_of_interest)
    
    print(paste("xgboost done. Plotting", gene_of_interest))
    # Plot the expression of the gene of interest
    plot_gene <- plot_expression(sce.enhanced, gene = gene_of_interest, title = sample, expr_limits = gene_limits)
} else {
    # If gene data is available, proceed to plot
    print(paste("plotting", gene_of_interest))
    plot_gene <- plot_expression(sce.enhanced, gene = gene_of_interest, title = sample, expr_limits = gene_limits)
}

# Print the plot
print(plot_gene)
```
