```r
# use kernel python [conda env:ipykernel_py3]
import pandas as pd
import matplotlib.pyplot as plt
import arviz as az
import sccoda 
from sccoda.util import comp_ana as mod
from sccoda.util import cell_composition_data as dat
from sccoda.util import data_visualization as viz
print(sccoda.__version__)
```

```r
# Load data
cell_counts = pd.read_csv("/Routput/Seurat/scCODA/cluster_contribution_WM.csv")
cell_counts.rename( columns={'Unnamed: 0':'sample_ID'}, inplace=True )
```

```r
cell_counts['condition'] = ['CWM', 'CWM', 'NAWM', 'NAWM', 'NAWM', 'active', 'active', 'active', 'active',
                            'act/inact', 'act/inact','act/inact','act/inact','act/inact','act/inact']
cell_counts
```

```r
# Convert data to anndata object
spots_clusters = dat.from_pandas(cell_counts, covariate_columns=['sample_ID', 'condition'])
```

```r
viz.boxplots(spots_clusters, feature_name="condition")
plt.show()

viz.stacked_barplot(spots_clusters, feature_name="condition")
plt.show()
```

```r
viz.rel_abundance_dispersion_plot(
    data=spots_clusters,
    abundant_threshold=0.9)

plt.show()
```

```r
model_st = mod.CompositionalAnalysis(spots_clusters, formula="condition", reference_cell_type="5")
# or reference_cell_type = "automatic" 
st_results = model_st.sample_hmc()
st_results.summary()
```

```r
st_results.set_fdr(est_fdr=0.05)
st_results.effect_df.to_csv(outdir + 'WM-default-scCODA_ref_set_to_cluster5.csv')
st_results.credible_effects().to_csv(outdir + 'WM-default-credible-effect-scCODA_ref_set_to_cluster5.csv')
print(st_results.effect_df)
```
