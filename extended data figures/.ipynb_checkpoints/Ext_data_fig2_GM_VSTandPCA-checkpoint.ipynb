```r
#################################################################
# Author: Astrid Alsema
# Date: July 2020. GM - adapted by Marion in Aug 2021
# Dataset   : Visium Spatial Transcriptomics for MS lesions
# Purpose: GM neuronal layers, create a PCA plot using the DESeq2::vst variance stabilizing transformed counts

# Required inputs:
# - Pseudobulk_GM_v4.csv : containing count table with pseudobulk samples 
# - target_pseudobulk_GM_v4_subgroup.csv: csv file with information about samples
# Output: PCA plot with one color per neuronal layer
#################################################################
rm(list = ls())

library(vsn)
library(plyr)
library(dplyr)
library(DESeq2)
library(stringr)
require(ggplot2)
require(ggrepel)
require(RColorBrewer)

# load files
countfile <- read.csv(file = "pseudobulk/Pseudobulk_GM_v4.csv", row.names = 1)
# create coldata
target <- read.csv(file = "pseudobulk/target_pseudobulk_GM_v4_subgroup.csv", sep=",", header = T, 
                   stringsAsFactors = F) # 
head(target)


#should be true
identical(colnames(countfile), target$Unique_ID)

```

```r
# if previous step is TRUE continue
rownames(target) <- target$Unique_ID
target <- target[match(colnames(countfile), rownames(target)) , ]

# subset only neuronal GM layers + WM
to_drop <- target$Unique_ID[target$Subgroup %in% c("technical", "meninges", "leuko_lesion", "WM_lesion", "bloodvessel", "leukolesion", "layer_4_5_6", "layer_3_4", "not_clear")]
print(to_drop)
target.4 <- target[!target$Unique_ID %in% to_drop , ]
target.4 <- droplevels(target.4)
countfile.4 <- countfile[!colnames(countfile) %in% to_drop]
```

```r
table(target.4$pathology)
```

```r
####################################################################################################

# make DESEQ2 object
dds0 <- DESeqDataSetFromMatrix(countData = as.matrix(countfile.4),
                               colData = target.4,
                               design = ~ Subgroup)
dds0$Group <- relevel( dds0$Subgroup, "WM" )

# # pre-filtering, remove genes with less than 2 reads. Relevant for PCA. many zero counts still in ST data (compared to regular bulk),
keep <- rowSums(counts(dds0)) >= 2 #
table(keep)
dds <- dds0[keep,]
# many genes will be removed, since there are still many zero counts in this data.
print(paste("we prefiltered out", nrow(dds0) - nrow(dds), "lowly expressed genes"))
print(paste("we keep", nrow(dds), "genes for PCA"))

```

```r
########################################## Normalize for PCA #########################################################
vsd <- vst(dds, blind = FALSE, fitType = "parametric") # variance stabilizing tranformation, similar to SCtransform
meanSdPlot(assay(vsd), ranks = FALSE) # vsd make the most straight line, also for medium abundance genes. compared to rlog
```

```r
########################################## compute PCs #########################################################
# # calculate size Factor to correct library size in DE analysis
dds <- estimateSizeFactors(dds, type = "ratio") # median ratio method, because Stahl et al used this for mouse olfactory pseudobulk, ST kit.

# use vsd counts for pca
transformed.counts <- as.matrix(assay(vsd))
# PCA
ntop = 500
rv <- rowVars(transformed.counts)
select <- order(rv, decreasing = TRUE)[seq_len(min(ntop, length(rv)))]
pca <- prcomp(t(transformed.counts[select, ]))
percentVar <- pca$sdev^2/sum(pca$sdev^2)
```

```r

# define the colors
mycolors <- rev(c("#82C56E", "#F4CC76", "#92B0DC", "#5EC6CA", "#F79B77", "#C4A6CF", "#9B9EA0"))

# subset lesions vs control
group <- factor(dds$Subgroup, levels = c("WM", "layer_6", "layer_5","layer_4", "layer_3","layer_2", "layer_1", "layer_4_5_6", "meninges", "layer_3_4", "bloodvessel", "leukolesion", "WM_lesion"))
sample_ID <- factor(dds$Sample_ID) # , levels = c("ST55", "ST56", "ST57", "ST58", "ST59", "ST60", "")
sex <-factor(dds$Seks,levels=c("f","m"))
pathology <-factor(dds$pathology, levels=c("CGM", "NAGM", "subpial", "leukocortical"))

# ggplot friendly dataframe d
d <- data.frame(PC1 = pca$x[, 1], PC2 = pca$x[, 2], group = group, donor = sample_ID, sex = sex) # here you can change which PC you want to plot

# make a plot
ggplot(data = d, aes_string(x = "PC1", y = "PC2", color = "group",shape="sex")) +
  geom_point(aes(), size = 6, alpha = 1 ) +
  xlab(paste0("PC1: ", round(percentVar[1] * 100), "% variance")) +
  ylab(paste0("PC2: ", round(percentVar[2] * 100), "% variance")) +
  theme_classic() +
  labs(title = "ST pseudobulk", subtitle = "") +
  scale_color_manual(values = mycolors) + 
  scale_shape_manual(values = seq(15,19))+
  scale_x_continuous(limits = c(-55, 55)) +
  scale_y_continuous(limits = c(-55, 55)) +
  theme(aspect.ratio = 1)

```
