```r
#subset a sample 
ST <- subset(datasets, subset = orig.ident == name)
head(ST)
print(paste("analysing", ST$orig.ident[1]))
ST_ID <- name

#load .csv file with specific groups (rim, PL-WM, core)
group_info <- read.csv(paste0("/data/bcn/Datasets/ST_WM/RData/persample/rim_core/excl_GM/", ST_ID, "_barcodes_rimcoregroup.csv"), 
                        header = TRUE,
                        row.names = 1)
head(group_info)

 group_IDs <- unique(group_info$group)
    print("available spot groups:")
    print(table(group_info$group))
    
    
```

```r
#do this for each sample
# calculate BV index 

bloodvessel_tmp <- data.frame(matrix(ncol = 3, nrow = length(group_IDs)))
colnames(bloodvessel_tmp) <- c("sample_ID", "sub_group", "BV_index")

for (g in 1:length(group_IDs)){
        # define the group
        current_group <- group_IDs[g]
        spots_group <- group_info[group_info$group == current_group , 1 ]
    
                
        # subset spots belonging to one group
        print("filtering spots from group")
        print(current_group)
        STsub <- subset(ST, cells = spots_group) 
        total_spots <- ncol(STsub) #number
        
        
        BB.spots <- sum(STsub$Inflamed_BV_threshold)
        index <- (BB.spots/total_spots)*100
        bloodvessel_tmp[g,1] <- ST_ID
        bloodvessel_tmp[g,2] <- current_group
        bloodvessel_tmp[g,3] <- index
#         colnames(bloodvessel_tmp) <- c("sample_ID", "sub_group", "BV_index")
        # calculate the number of pos. spots / total spots per group  
    }
print(bloodvessel_tmp)
bloodvessel_index_ST68 <- bloodvessel_tmp
bloodvessel_index_ST68

bloodvessel_index_ST68$sub_group <- substring(bloodvessel_index_ST68$sub_group, 6)
bloodvessel_index_ST68
```

```r
# combine indexes of all samples into one file
index_allsamples <- rbind (bloodvessel_index_ST33, bloodvessel_index_ST67,bloodvessel_index_ST68, bloodvessel_index_ST69,bloodvessel_index_ST70, bloodvessel_index_ST71, bloodvessel_index_ST72, bloodvessel_index_ST73, bloodvessel_index_ST74, bloodvessel_index_ST79)
index_allsamples


write.csv(index_allsamples, "index_active-mixed.csv")
```
