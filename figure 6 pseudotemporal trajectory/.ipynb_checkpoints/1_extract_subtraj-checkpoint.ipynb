```r
########################################################################
# Author    : A. Alsema
# Date      : May 2022
# Dataset   : Visium Spatial Transcriptomics for MS lesions, 15 slices with WM 
# Purpose   : extract branches in trajectory to further investigate as subtrajectory using monocle3::choose_graph_segments 

# Required Inputs:   
# - cds object including total trajectory and all genes: "3.monocle_traj_WM_allgenes.rds"
# - indir: path to cds object 
# - end node names per subtrajectory, e.g. Y_129

# Output    : 
# - cds_subset per subtrjactory, e.g. "4-4.startnode_to_rims-microglia.rds"
########################################################################
```

```r
rm(list = ls())
library(monocle3)
library(ggplot2)
library(dplyr)
library(Matrix)

indir = "RData/monocle3/"
```

### load data

```r
cds <- readRDS(paste0(indir, "3.monocle_traj_WM_allgenes.rds"))
dim(cds)
```

### investigate the end nodes

```r
plot_cells(cds, label_principal_points = TRUE,  
           graph_label_size = 3, color_cells_by = "pseudotime")
```

```r
spots_high <- cds[ , colData(cds)$Pseudotime > 20] #subset spots with high pseudotime
```

```r

plot_cells(spots_high, label_principal_points = TRUE, 
           graph_label_size = 3, color_cells_by = "pseudotime")
```

conclusion: it would be interesting to zoom in on end-points with high pseudotime Y_129 and Y_17
Also, Y_115 is interesting as it corresponds to the UMAP area enriched in lesion rims and microglia genes. Subset three subtrajectories to investigate.

## Extract subtrajectory 1: 
### Pseudotime trajectory to the microglia & rim enriched UMAP area. 
ending in node Y_115

```r
cds_subset <- choose_graph_segments(cds, starting_pr_node = "Y_42", 
                                         ending_pr_nodes = c("Y_115"),
                                         clear_cds = FALSE)

plot_cells(cds_subset, label_principal_points = TRUE,  graph_label_size = 3, color_cells_by = "pseudotime")

dim(cds_subset)
```

```r
saveRDS(cds_subset, file = paste0(indir, "4-4.startnode_to_rims-microglia.rds"))
```

## Extract subtrajectory 2: 
### Pseudotime trajectory to the UMAP area enriched in plasma-cell genes & active core.
ending in node Y_17

```r
cds_subset <- choose_graph_segments(cds, starting_pr_node = "Y_42", 
                                         ending_pr_nodes = c("Y_17", "Y_138", "Y_22"),
                                         clear_cds = FALSE)

plot_cells(cds_subset, label_principal_points = TRUE,  graph_label_size = 3, color_cells_by = "pseudotime")

dim(cds_subset)
```

```r
saveRDS(cds_subset, file = paste0(indir, "4-1.startnode_to_infl.rds"))
```

## Extract subtrajectory 3: 
### Pseudotime trajectory to the UMAP area enriched in mixed act/inact lesion core.
ending in node Y_129

```r
cds_subset <- choose_graph_segments(cds, starting_pr_node = "Y_42", 
                                         ending_pr_nodes = c("Y_129", "Y_138", "Y_22"),
                                         clear_cds = FALSE)
```

```r
plot_cells(cds_subset, label_principal_points = FALSE, color_cells_by = "pseudotime",
                   show_trajectory_graph=TRUE,
                   label_groups_by_cluster=FALSE,
                   label_leaves=FALSE,
                   label_roots = TRUE,
                   label_branch_points=FALSE,
                   trajectory_graph_segment_size = 1.5,
                   graph_label_size= 5) + 
                        theme_void() +
                        theme( legend.position = "none")
```

```r
saveRDS(cds_subset, file = paste0(indir, "4-2.startnode_to_active_mixed.rds"))
```
